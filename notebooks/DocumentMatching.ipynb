```python
import torch
from transformers import AutoModel, AutoTokenizer
import re

tokenizer = AutoTokenizer.from_pretrained("vinai/phobert-base")
phobert = AutoModel.from_pretrained("vinai/phobert-base", output_hidden_states = True)

```

```python
def process_input(doc_path):
    vi_doc = []
    km_doc = []
    regex = '[àáãạảăắằẳẵặâấầẩẫậèéẹẻẽêềếểễệđìíĩỉịòóõọỏôốồổỗộơớờởỡợùúũụủưứừửữựỳỵỷỹýÀÁÃẠẢĂẮẰẲẴẶÂẤẦẨẪẬÈÉẸẺẼÊỀẾỂỄỆĐÌÍĨỈỊÒÓÕỌỎÔỐỒỔỖỘƠỚỜỞỠỢÙÚŨỤỦƯỨỪỬỮỰỲỴỶỸÝ]'

    with open(doc_path, encoding='utf-8') as f:
        vi_sen = []
        km_sen = []
        lines = f.readlines()
        for line in lines:
            if len(re.findall(regex, line)) > 0: 
                vi_sen.append(line)
                if len(km_sen) > 0:
                    km_doc += km_sen
                    km_sen = []
            else:
                vi_doc += vi_sen
                vi_sen = []
                km_sen.append(line)
        if len(vi_sen) > 0: vi_doc += vi_sen
        if len(km_sen) > 0: km_doc += km_sen
    return vi_doc, km_doc
path = './samples.txt'
vi_doc, km_doc = process_input(path)
with open('samples_vi.txt','w', encoding='utf-8') as f:
    for line in vi_doc:
        f.write(line)

with open('samples_km.txt', 'w', encoding='utf-8') as f:
    for line in km_doc:
        f.write(line)
# for vi_sen, km_sen in zip(vi_doc, km_doc):
#     print(vi_sen)
#     print(km_sen)

```

```python

# tokenize sentences
def tokenize_sentences(doc):
    encoded_sentences = tokenizer(doc, padding=True, truncation=True, return_tensors='pt')
    return encoded_sentences['input_ids'], encoded_sentences['attention_mask']



```

```python
trans_doc_ids, trans_doc_mask = tokenize_sentences(trans_doc)
vi_doc_ids, vi_doc_mask = tokenize_sentences(vi_doc)
print(trans_doc_ids.shape)
print(vi_doc_ids.shape)
```

```python
def decode_sentence(sen_encoding):
    for word in sen_encoding:
        print(tokenizer.decode([word]))
```

```python

# feed into model and get sentence embedding
def get_sentence_embedding(doc_ids, doc_mask):
    # res = []
    for ids, mask in zip(doc_ids, doc_mask):
        sentence_embed = phobert(ids.reshape(1, -1), attention_mask=mask.reshape(1, -1))
        yield sentence_embed



# print(len(trans_doc_final_embedding))
# print(len(vi_doc_final_embedding))
#
# print(trans_doc_embedding[0].shape)
# print(vi_doc_ids[0].shape)
```

```python
# store result in a generator
trans_doc_embedding = get_sentence_embedding(trans_doc_ids, trans_doc_mask)
vi_doc_embedding = get_sentence_embedding(vi_doc_ids, vi_doc_mask)

trans_doc_final_embedding = []
vi_doc_final_embedding = []

for sentence_embedding in trans_doc_embedding:
    #print(sentence_embedding[0].shape)
    trans_doc_final_embedding.append(sentence_embedding)
for sentence_embedding in vi_doc_embedding:
    #print(sentence_embedding[0].shape)
    vi_doc_final_embedding.append(sentence_embedding)
```

```python
def concatenate_hidden_states(embed):
    return torch.cat(embed[2][-4:], axis = 1)
```

```python
def cosine_similarity(embed1, embed2):
    return torch.dot(embed1.reshape(768), embed2.reshape(768))/(torch.norm(embed1) * torch.norm(embed2))

def compare_similarity(doc1, doc2):
    match_table = []
    for i, sen1 in enumerate(doc1):
        match_table.append([])
        for j, sen2 in enumerate(doc2):
            # 4 last layers
            similarity = cosine_similarity(torch.mean(concatenate_hidden_states(sen1), dim = 1), torch.mean(concatenate_hidden_states(sen2), dim = 1))
            # last layer
            # similarity = cosine_similarity(torch.mean(sen1[0], dim = 1), torch.mean(sen2[0], dim = 1))
            # output of 
            #similarity = cosine_similarity(sen1[1], sen2[1])
            match_table[i].append(similarity)
    return match_table

```

```python
match_table = compare_similarity(trans_doc_final_embedding, vi_doc_final_embedding)
for i in range(len(match_table)):
    for j in range(len(match_table[i])):
        print('Translate ' + str(i) + ': ' + trans_doc[i])
        print('Original ' + str(j) + ': ' + vi_doc[j])
        print('Score: ')
        print(match_table[i][j])
        print('-'*50)
```

```python
for i in range(len(match_table)):
    if (match_table[i][i] < 0.8):
        print('Original: ' + vi_doc[i])
        print('Original decode: ')
        decode_sentence(vi_doc_ids[i])
        print('Translate: ' + trans_doc[i])
        print('Translate decode: ')
        decode_sentence(trans_doc_ids[i])
        print('Score for pair ' + str(i) + ':')
        print(match_table[i][i])
        print('-'*50)
```

```python
def segment_sentence(sen):
    word_segmented_sen = annotator.tokenize(sen)
    res = ''
    for sen in word_segmented_sen:
        for word in sen:
            res += word + ' '
    return res 
```
